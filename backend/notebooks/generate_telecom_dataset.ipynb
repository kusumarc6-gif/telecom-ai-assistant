```python
from faker import Faker
import pandas as pd
import random
from datetime import datetime, timedelta

fake = Faker("en_IN")
random.seed(42)
```

```python
NUM_CUSTOMERS = 5000
```

```python
customers = []

connection_types = ["Prepaid", "Postpaid"]
plan_names = ["Basic 199", "Unlimited 399", "Family 699", "Premium 999"]
statuses = ["Active", "Inactive", "Suspended"]

for i in range(1, NUM_CUSTOMERS + 1):
    customers.append({
        "customer_id": f"CUST{i:05d}",
        "first_name": fake.first_name(),
        "last_name": fake.last_name(),
        "mobile_number": "9" + "".join(str(random.randint(0, 9)) for _ in range(9)),
        "email": fake.email(),
        "city": fake.city(),
        "state": fake.state(),
        "connection_type": random.choice(connection_types),
        "plan_name": random.choice(plan_names),
        "account_status": random.choice(statuses),
        "created_at": fake.date_time_between(start_date="-5y", end_date="now")
    })

customer_df = pd.DataFrame(customers)

customer_df.head()
```

```python
customer_df.to_csv("../data/customer_master.csv", index=False)

print("customer_master.csv created successfully!")
```

```python
billing_data = []

payment_statuses = ["Paid", "Paid", "Paid", "Pending", "Overdue"]

for customer in customer_df.itertuples(index=False):

    for month in range(12):   # 12 months of billing

        bill_month = fake.date_between(start_date="-12M", end_date="today")

        if customer.plan_name == "Basic 199":
            amount = 199
        elif customer.plan_name == "Unlimited 399":
            amount = 399
        elif customer.plan_name == "Family 699":
            amount = 699
        elif customer.plan_name == "Premium 999":
            amount = 999
        else:
            amount = random.choice([199,299,399,499,699,999])

        late_fee = random.choice([0,0,0,25,50,100])

        payment_status = random.choice(payment_statuses)

        billing_data.append({
            "customer_id": customer.customer_id,
            "bill_month": bill_month,
            "bill_amount": amount,
            "due_date": bill_month + timedelta(days=15),
            "payment_date": bill_month + timedelta(days=random.randint(0,20)),
            "payment_status": payment_status,
            "late_fee": late_fee,
            "total_amount": amount + late_fee
        })

billing_df = pd.DataFrame(billing_data)

billing_df.head()
```

```python
service_requests = []

request_types = [
    "SIM Activation",
    "Broadband Installation",
    "Plan Change",
    "Network Issue",
    "Bill Dispute",
    "Address Update",
    "Fiber Upgrade",
    "SIM Replacement",
    "Service Cancellation",
    "New Connection"
]

statuses = ["Open", "In Progress", "Resolved", "Closed"]

priorities = ["Low", "Medium", "High", "Critical"]

agents = [
    "Rahul Sharma",
    "Priya Reddy",
    "Amit Kumar",
    "Sneha Patel",
    "Arjun Singh",
    "Neha Verma",
    "Rohit Rao",
    "Kiran Das"
]

for customer in customer_df.itertuples(index=False):

    for _ in range(random.randint(1,4)):

        request_date = fake.date_time_between(start_date="-2y", end_date="now")

        status = random.choice(statuses)

        resolution_date = (
            request_date + timedelta(days=random.randint(1,10))
            if status in ["Resolved","Closed"] else None
        )

        service_requests.append({
            "customer_id": customer.customer_id,
            "request_type": random.choice(request_types),
            "request_date": request_date,
            "status": status,
            "priority": random.choice(priorities),
            "assigned_to": random.choice(agents),
            "resolution_date": resolution_date
        })

service_df = pd.DataFrame(service_requests)

service_df.head()
```

```python
service_df.to_csv("../data/service_requests.csv", index=False)

print("service_requests.csv created successfully!")
```

```python
complaints = []

categories = [
    "Network Issue",
    "Billing Issue",
    "Slow Internet",
    "Call Drop",
    "SIM Activation",
    "Recharge Failure",
    "Fiber Issue",
    "Plan Change",
    "Service Quality"
]

statuses = ["Open", "In Progress", "Resolved", "Closed"]

for request_id, request in enumerate(service_df.itertuples(index=False), start=1):

    if random.random() < 0.6:

        complaint_date = request.request_date

        resolved_date = (
            complaint_date + timedelta(days=random.randint(1, 7))
            if request.status in ["Resolved", "Closed"] else None
        )

        complaints.append({
            "customer_id": request.customer_id,
            "request_id": request_id,
            "complaint_category": random.choice(categories),
            "complaint_description": fake.sentence(nb_words=10),
            "complaint_date": complaint_date,
            "status": request.status,
            "resolution_notes": fake.sentence(nb_words=12) if resolved_date else None,
            "resolved_date": resolved_date
        })

complaint_df = pd.DataFrame(complaints)

complaint_df.head()
```
